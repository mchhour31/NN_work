```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = 40 #Hz
signal = lambda t: np.cos(2*np.pi*t) + np.cos(2*np.pi*t*f)
```

```python
def signal(t):
    return np.cos(2*np.pi*t) + np.cos(2*np.pi*t*f)
```

```python
tmin = -0.3
tmax = 0.3
N = int((tmax-tmin)*800)
t = np.linspace(tmin, tmax, N)
x = signal(t)
plt.plot(t,x)
plt.xlabel('time / s')
plt.ylabel('signal / a.u.')
```

```python
T_N = 1 / (2*f)
N_s = int(np.round((tmax-tmin)/T_N))+1
```

```python
t_s = np.linspace(tmin, tmax, N_s)
t_s
```

```python
x_s = signal(t_s)
```

```python
x_s
```

```python
plt.plot(t,x)
plt.plot(t_s, x_s, 'bo')
plt.xlabel('time / s')
plt.ylabel('signal / a.u.')
```

```python
from scipy.signal import resample
```

```python
x_r = resample(x_s[:-1], t.shape[0])
```

```python
plt.plot(t,x)
plt.plot(t, x_r, 'r:')
plt.xlabel('time / s')
plt.ylabel('signal / a.u.')
```
